Unicode 把所有字符分为 17 个“平面”：

- **BMP (Basic Multilingual Plane)**：U+0000–U+FFFF
  - 常见英文、拉丁字母、中文、日文、韩文、常用符号
- **补充平面 (Supplementary Planes)**：U+10000–U+10FFFF
  - emoji 😀😂👍
  - 古文字、特殊乐谱、冷门符号

```python
matches = merged['Description'].dropna().str.findall(r'[\U00010000-\U0010FFFF]') 
for i, m in enumerate(matches):
    if m:  # Only print rows with matches
        print(f'Row {i}: {m}')
```

```python
matches = merged['Description'].dropna().str.findall(r'[^\x00-\x7F]+') # Match non-ASCII characters, that is non-English characters
for i, m in enumerate(matches):
    if m:  # Only print rows with matches
        print(f'Row {i}: {m}')
```

```python
import re
def clean_text(text):
    if pd.isna(text):
        return 'NaN'
    # Convert to lowercase
    text = text.lower()
    # Remove XML/JSON tags, emojis, and non-UTF-8 characters
    text = re.sub(r'<[^>]*>|[\U00010000-\U0010FFFF]', '', text)
    # Remove non-English characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text
# Apply cleaning to the specified columns
columns_to_clean = ['Title', 'City', 'Country', 'Tags', 'Description']
for col in columns_to_clean:
    if col in merged.columns:
        merged[col] = merged[col].apply(clean_text)
```

```python
merged[['City', 'Country', 'Description']] = merged[['City', 'Country', 'Description']].replace(r'^\s*$', "NaN", regex=True)
```

```python
merged
```

```python
columns_to_convert = ['Post_ID', 'Server', 'Is_Public', 'Is_Friend', 'Is_Family', 'Farm']
merged[columns_to_convert] = merged[columns_to_convert].map(
    lambda x: str(int(x)) if not pd.isna(x) else x
)
```

```python
merged.fillna('NaN', inplace=True)
merged
```
