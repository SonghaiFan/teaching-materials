## **Week 3 Applied Session: Regular Expression**

In the lecture, you have learnt how to use regular expressions. To quickly review some regular expression syntax:


* <font color="red">[0-9]</font> Matches a single digit
* <font color="red">[a-z0-9]</font> Matches a single character that must be a lower case letter or a digit.
* <font color="red">[A-Za-z]</font> Matches a single character that much be a upper/lower case letter
* <font color="red">\d</font> Matches any decimal digit; equivalent to the set [0-9].
* <font color="red">\D</font> Matches characters that are not digits, which is equivalent to [^0-9] or [^\d].
* <font color="red">\w</font> Matches any alphanumeric character, which is equivalent to [a-zA-Z0-9].
* <font color="red">\W</font> Matches any non-alphanumeric character; which is equivalent to [^a-zA-Z0-9] or [^\w].
* <font color="red">\s</font> Matches any whitespace character; which is equivalent to [\t\n\r\f\v], where \t indicates taps, \n  line feeds, \r carriage returns, \f form feeds and \v vertical tabs.
* <font color="red">\S:</font> Matches any non-whitespace character; which is equivalent to  [^ \t\n\r\f\v].
* <font color="red">ˆ</font> Matches the start of the line.
* <font color="red">$</font> Matches the end of the line.
* <font color="red">.</font> Matches any character (a wildcard).
* <font color="red">*</font> Matches when the preceding character occurs zero or more times
* <font color="red">?</font> Matches when the preceding character occurs zero or one times
* <font color="red">+</font> Matches when the preceding character occurs one or more times

More information can be found here :

## **2. Regular Expression for Dates**

Refine the regular expression for date to distinguish months with 29/30/31 days.

**Note**: assume all years are a leap year, which means every Feburary has 29 days.

````python
# define a function to validate the input date
def date(pattern, m):
    if re.match(pattern, m):
        print (m + " is a date")
    else:
        print (m + " is NOT a date")
````

````python
regex = r'''(?x)
    (?:
    # February (29 days every year)
      ([12][0-9]|0?[1-9])[/-](0?2)
    # 30-day months
      |(30|[12][0-9]|0?[1-9])[/-](0?[469]|11)
    # 31-day months
      |(3[01]|[12][0-9]|0?[1-9])[/-](0?[13578]|1[02])
    )
    # Year
    [/-]((?:[0-9]{2})?[0-9]{2})
'''
````

````python
#date(r"((31[/-](0?[13578]|1[02]))|(30[/-](0?[469]|11))|(28[/-]02))[/-]((?:\d{2})?\d{2})", "28/02/2019")
date(regex, "28/02/2019")
date(regex, "31/04/2019")
date(regex, "29/05/2019")
date(regex, "31/06/2019")
````

### Parse raw Air Crashes data with regular expressions

We'll now convert the unstructured `AirCrashes_raw.csv` (really a line-oriented text file) into a structured DataFrame.

Parsing logic:
1. Detect header lines starting with `Incident` to start a new record.
2. Extract Flight (flight identifiers part), Mode (aircraft type(s)), and Year from the header using regex.
3. Subsequent non-empty lines are `Label,Value` pairs. Some values may be empty (e.g. `Crew,`).
4. Blank line or a new `Incident` line finalises the current record.
5. Post-process: clean numeric fields, validate dates, flag suspicious years or malformed dates.

We'll build a parser function so you can reuse / adapt it. After parsing we'll get a DataFrame similar to the previously provided structured `AirCrashes.csv`. Next cell implements this.

````python
# 步骤1：按纯文本逐行读取原始文件（而不是直接让 pandas 解析成一列）
# 这样可以精细控制“事件头行（Incident ...）”与“字段行（Label,Value）”以及空行的识别。
from pathlib import Path  # Path 对象便于处理文件路径
raw_path = Path('AirCrashes_raw.csv')  # 指向原始数据文件
raw_text = raw_path.read_text(encoding='utf-8')  # 读取整个文件为一个长字符串
lines = raw_text.splitlines()  # 分割成行列表（保留顺序）
print(f'Total lines (including blanks): {len(lines)}')  # 打印总行数，包含空行用于调试
print('First 8 lines:')
for l in lines[:8]:  # 预览前 8 行，确认格式：第一行为 Incident 开头，之后是若干 “字段,值” 行
    print(l)
````

#### Step 2: Identify incident header lines
We first locate lines that begin a new incident. Pattern to capture:
`Incident <description> in <year>`
We'll test the regex against a sample of lines.

````python
import re  # 导入正则表达式库
# incident_re 用于捕获“事件头行”中间部分(航班/飞机描述)与年份
# ^Incident\s+ : 行首以 Incident 开始，后跟一个或多个空白
# (.+?)         : 非贪婪匹配中间描述（航班 + 'involving' + 机型等）
# \s+in\s+     : 关键字 in，两侧若干空白
# (\d{4})$     : 4 位年份，行尾结束
incident_re = re.compile(r'^Incident\s+(.+?)\s+in\s+(\d{4})$')
# 采样部分以 Incident 开头的行做测试
header_candidates = [l for l in lines if l.startswith('Incident')][:5]
for hc in header_candidates:
    m = incident_re.match(hc)
    # 验证匹配是否成功，并输出捕获到的组（中间描述, 年份）
    print(hc, '->', 'MATCH' if m else 'NO MATCH', m.groups() if m else '')
````

#### Step 3: Split header into `Flight` and `Mode`
We now take the portion between `Incident` and the year, and split around the keyword `involving` to separate the flight identifiers from aircraft type(s).

````python
# header_split_re: 再次匹配“Incident ... in 年”，便于拆出中段描述与年份
header_split_re = re.compile(r'^Incident\s+(?P<mid>.+?)\s+in\s+(?P<Year>\d{4})$')
# flight_mode_re: 将中段描述再按关键字 involving 拆分为 Flight 与 Mode
# ^(?P<Flight>.+?)\s+involving\s+(?P<Mode>.+)$
#  - (?P<Flight>.+?) 非贪婪匹配航班部分（可能含多个航班）
#  - involving 关键字
#  - (?P<Mode>.+)   剩余全部作为机型/型号描述
flight_mode_re = re.compile(r'^(?P<Flight>.+?)\s+involving\s+(?P<Mode>.+)$', re.IGNORECASE)

examples = [l for l in lines if l.startswith('Incident')][:3]
for ex in examples:
    m = header_split_re.match(ex)
    if m:
        mid = m.group('mid')  # 中间部分（含 Flight + involving + Mode）
        year = m.group('Year')
        # 为复用 flight_mode_re，将 mid 前面补一个单词 Incident，便于模式统一
        fm = flight_mode_re.match('Incident ' + mid)
        if fm:
            # 去掉前置 Incident，strip 清理多余空格
            flight = fm.group('Flight').replace('Incident','').strip()
            mode = fm.group('Mode').strip()
        else:
            # 如果没有 involving，就把整个 mid 当做 Flight
            flight, mode = mid, None
        print(f'Original: {ex}\n  Flight: {flight}\n  Mode: {mode}\n  Year: {year}')
````

#### Step 4: Parse label,value lines
Each subsequent non-empty line (until a blank) has the form `Label,Value`. We test a regex and handle empty values (e.g., `Crew,`).


## 1. 什么是捕获组
在正则表达式里，用小括号 `(...)` 括起来的部分叫 **捕获组**。  
它的作用是：
- 把匹配结果的一部分“捕获”下来，以便后续提取或复用。  
- 捕获组默认用编号表示，编号从 1 开始（组 0 是整个正则）。  

例子：
```python
import re
m = re.match(r'(\d{4})-(\d{2})-(\d{2})', "2025-08-17")
print(m.group(1))  # 2025
print(m.group(2))  # 08
print(m.group(3))  # 17
```

---

## 2. 命名捕获组
**语法**：
```
(?P<name>pattern)
```
- `name` 是组的名字（必须是合法的标识符：字母/数字/下划线，不能以数字开头）。  
- `pattern` 是要匹配的内容。  

这样你就可以用名字而不是数字来访问结果，更清晰。  

例子：
```python
import re
m = re.match(r'(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})', "2025-08-17")

print(m.group("year"))   # 2025
print(m.group("month"))  # 08
print(m.group("day"))    # 17

print(m.groupdict())     # {'year': '2025', 'month': '08', 'day': '17'}
```


````python
# label_value_re 捕获形如 Label,Value 的行
# (?P<label>[^,]+?) : 逗号前的字段名（非逗号最短匹配）
# ,
# (?P<value>.*)     : 逗号后的全部内容（可以为空字符串）
label_value_re = re.compile(r'^(?P<label>[^,]+?),(?P<value>.*)$')
# 在首个 Incident 后收集若干字段行作为示例
sample_label_lines = []
collect = False
for l in lines:
    if l.startswith('Incident'):
        collect = True
        continue
    if collect:
        if not l.strip():  # 空行则结束示例采样
            break
        if ',' in l:
            sample_label_lines.append(l)
    if len(sample_label_lines) >= 10:
        break
for sl in sample_label_lines:
    m = label_value_re.match(sl)
    print(sl, '->', m.group('label'), '|', m.group('value'))
````

#### Step 5: Assemble records
Iterate through lines, start a new dict on each header, accumulate label/value pairs, flush on blank line. We'll print the first parsed record for inspection.

````python
# 迭代所有行并组装成 records 列表；每条记录是一个字典
records = []  # 存放所有事故记录
current = {}  # 当前正在填充的记录

for line in lines + ['']:  # 结尾额外加一个空字符串，确保最后一条能被 flush
    line = line.strip('\ufeff').rstrip()  # 去除 BOM 和末尾空白
    if not line:  # 空行表示一个事件块结束
        if current.get('Flight') or current.get('Year'):
            records.append(current)
            current = {}
        continue
    mh = incident_re.match(line)
    if mh:  # 事件头行，开始新记录
        if current.get('Flight') or current.get('Year'):
            records.append(current)
            current = {}
        header_mid, year = mh.groups()
        fm = flight_mode_re.match('Incident ' + header_mid)
        if fm:  # 成功分离 Flight 与 Mode
            current['Flight'] = fm.group('Flight').replace('Incident','').strip()
            current['Mode'] = fm.group('Mode').strip()
        else:
            current['Flight'] = header_mid
        current['Year'] = int(year)  # 年份转为整数
        continue
    mv = label_value_re.match(line)  # 普通字段行
    if mv:
        lab, val = mv.group('label').strip(), mv.group('value').strip() or None
        current[lab] = val  # 若值为空串则存 None

print('Total records parsed (raw, before cleaning):', len(records))
print('First record keys:', list(records[0].keys()))
print('First record sample:', records[0])
````

#### Step 6: Create DataFrame and clean
Rename columns, convert numeric fields, validate date format, attempt minor date fix for obvious month/day swaps.

````python
import pandas as pd  # 数据框处理
parsed_df = pd.DataFrame(records)  # 将列表字典转为 DataFrame
print('Shape before cleaning:', parsed_df.shape)  # 查看初始维度

# 重命名列：Total Dead -> Total_Dead 便于无空格字段名
parsed_df = parsed_df.rename(columns={'Total Dead':'Total_Dead'})

# 数值列转换：提取可能混杂的纯数字（这里简单使用首个整数匹配）
for col in ['Total_Dead','Crew','Passengers','Ground','Year']:
    if col in parsed_df.columns:
        parsed_df[col] = (parsed_df[col].astype(str)
                           .str.extract(r'(-?\d+)', expand=False)  # 提取数字部分
                           .astype('Int64'))  # 使用 pandas 可空整型 Int64

# 经纬度列转为浮点，非法值 -> NaN
for col in ['Latitude','Longitude']:
    if col in parsed_df.columns:
        parsed_df[col] = pd.to_numeric(parsed_df[col], errors='coerce')

# 日期合法性初步校验 + 简单“月日调换”修正（当月>12 且日<=12 时交换）
import re as _re
fix_pattern = _re.compile(r'^(\d{4})-(\d{2})-(\d{2})$')

def fix_date(d):
    if pd.isna(d):
        return d
    m = fix_pattern.match(d)
    if not m:
        return d
    y, mth, day = m.groups()
    if int(mth) > 12 and int(day) <= 12:  # 识别明显调换
        return f"{y}-{day.zfill(2)}-{mth.zfill(2)}"
    return d

parsed_df['Date'] = parsed_df['Date'].apply(fix_date)
parsed_df['Date_valid'] = parsed_df['Date'].str.match(r'^\d{4}-\d{2}-\d{2}$')
print('Invalid dates count:', (~parsed_df['Date_valid']).sum())

parsed_df.head()
````

#### Step 7: Reorder columns & (optionally) compare
Arrange columns to align with target structure and briefly compare with the preprocessed `aircrash` DataFrame if loaded earlier.

````python
# 设定期望列顺序（尽量与已有结构化 csv 对齐）
preferred_order = ['Flight','Mode','Casualties','Circumstances','Country','Crew','Date','Ground','Latitude','Location','Longitude','Notes','Passengers','Phase','Reason','Total_Dead','Type','Year']
# 先放置存在于 parsed_df 中且在 preferred_order 里的列，再追加剩余列
ordered_cols = [c for c in preferred_order if c in parsed_df.columns] + [c for c in parsed_df.columns if c not in preferred_order]
final_df = parsed_df[ordered_cols]
print('Final shape:', final_df.shape)

# 如果之前加载过结构化的 aircrash DataFrame，可做简单比较
if 'aircrash' in globals():
    print('Existing structured shape:', aircrash.shape)
    print('Columns missing vs target:', set(aircrash.columns) - set(final_df.columns))

final_df.head(10)
````
