### 1.1 Structure Validation

The first thing we can check is the structure of the data. We can check the following:

- The number of rows and columns
- The column names
- The data types of each column
- The number of missing values in each column
- The number of unique values in each column
- The number of duplicate rows

```python
# 1. the number of rows in the dataset
print("The number of rows in the dataset is: ", health_data.shape[0])
# 2. the number of columns in the dataset
print("The number of columns in the dataset is: ", health_data.shape[1])
# 3. the data types of each column
print("The data types of each column are: ", health_data.dtypes)
```

```python
# 4. the number of missing values in each column
print("The number of missing values in each column are: ", health_data.isnull().sum())
```

```python
# 5. the number of unique values in each column
print("The number of unique values in each column are: ", health_data.nunique())
```

```python
health_data.duplicated()
```

```python
# 6. the number of duplicate rows
print("The number of duplicate rows are: ", health_data.duplicated().sum())
# show the duplicate rows
print("The duplicate rows are: ", health_data[health_data.duplicated()])
```

Here we can see that the dataset has `534` rows of duplicate data. We can remove these duplicates to avoid any issues with the analysis.

```python
# remove the duplicate rows
health_data.drop_duplicates(inplace=True)
print("The number of rows after removing the duplicate rows is: ", health_data.shape[0])
```

### 1.2 Content Validation

The next thing we can check is the content of the data. We can check the following:

- The range of values for each column
- The distribution of values for each column
- The relationship between columns
- The consistency of values between columns

```python
# 1. the range of value for each column
print("The range of value for each column are: \n", health_data.describe().loc[['min', 'max']])
```

We can see that only the numerical columns are displayed in the above table.

```python
# 2. distribution of values for each column
print("The distribution of values for each column are: \n", health_data.describe())
```

We can find that the minimum `Billing Amount` value is negative (around -2008.492140). We can check the number of rows with negative `Billing Amount` values.

```python
# check the negative billing amount
print("The number of negative billing amount is: ", health_data[health_data['Billing Amount'] < 0].shape[0])
```

```python
# show the negative billing amount
print("The negative billing amount is: \n", health_data[health_data['Billing Amount'] < 0][['Billing Amount']])
```

<span style="color:red">How to deal with the negative billing amount values?</span>

Typically, the billing amount values are expected to be positive. However, there are instances where billing amounts can be negative. For instance, if a patient has `overpaid` their bill, the billing amount may appear as negative.

Similarly, when `refunding` money to the patient, the billing amount can be negative.

Additionally, hospitals may `adjust bills`, which can result in a negative billing amount.

<span style="color:red">Can we determine if any of the reasons mentioned above exist in this data?</span>

```python
# Lets check if there are any person with multiple records
print("The number of person with multiple records are: ", health_data['Name'].duplicated().sum())
# show the person with multiple records and negative billing amount
print("The person with multiple records and negative billing amount are: \n", health_data[(health_data['Name'].duplicated()) & (health_data['Billing Amount'] < 0)])
```

```python
negative_billing_amount_person = health_data[(health_data['Name'].duplicated()) & (health_data['Billing Amount'] < 0)]['Name'].unique()
print("The person with multiple records and negative billing amount are: \n", negative_billing_amount_person)
```

```python
# get one example
health_data[(health_data['Name'] == negative_billing_amount_person[0])]
```

```python
health_data[(health_data['Name'] == negative_billing_amount_person[1])]
```

__It seems that the 'Age' also has a mismatch issue regarding the above records__

### 1.3 Logical Validation

The last thing we can check is the logic of the data. This check involves checking the data against business rules or domain knowledge.

```python
# check if there is negative age
print("The number of negative age is: ", health_data[health_data['Age'] < 0].shape[0])
```

```python
# check the Blood Type column, and show the unique values to see if there is any value that is not in the specified values
print("The unique values in Blood Type column are: ", health_data['Blood Type'].unique())
```

```python
# check if there is any row Date of Admission is later than Date of Discharge
print("The number of rows where Date of Admission is later than Date of Discharge is: \n", health_data[health_data['Date of Admission'] > health_data['Discharge Date']].shape[0])
```
