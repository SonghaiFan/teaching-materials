## 16. Optional Solutions

先自己做，再运行下面单元格对答案。

```python
# Exercise 1A
products_solution = pd.DataFrame(
    {
        'product': ['pen', 'notebook', 'eraser'],
        'price': [2.5, 5.0, 1.2],
        'stock': [100, 50, 200],
    }
)

# Exercise 1B
scores_solution = pd.DataFrame(
    [[78, 85, 91], [88, 79, 84], [92, 90, 87]],
    columns=['math', 'english', 'science']
)

# Exercise 3A
high_score_students_solution = students[students['score'] > 80]

# Exercise 3B
target_students_solution = students[(students['major'] == 'IT') & (students['attendance'] >= 0.95)]

# Exercise 4A
selected_columns_solution = students[['name', 'major', 'score']]

# Exercise 4B
regex_columns_solution = students.filter(regex='a')

# Exercise 5A
sorted_students_solution = students.sort_values('score', ascending=False)

# Exercise 5B
renamed_students_solution = students.rename(columns={'attendance': 'attendance_rate'})

# Exercise 6A
sales_long_solution = pd.melt(
    sales_wide,
    id_vars='month',
    var_name='fruit',
    value_name='sales'
)

# Exercise 6B
sales_wide_again_solution = sales_long_solution.pivot(columns='fruit', values='sales', index='month').reset_index()

# Exercise 8A
missing_age_rows_solution = students[students['age'].isna()]

# Exercise 8B
students_dropna_age_solution = students.dropna(subset=['age'])
students_fillna_age_solution = students.assign(age=students['age'].fillna(students['age'].mean()))

# Exercise 9A
scores_with_total_solution = scores_detail.assign(
    total=lambda df: df['quiz'] + df['assignment'] + df['exam'],
    pass_=lambda df: (df['quiz'] + df['assignment'] + df['exam']) >= 50
)

# Exercise 9B
students_band_solution = students.assign(
    score_band=lambda df: np.where(
        df['score'] >= 85,
        'High',
        np.where(df['score'] >= 75, 'Medium', 'Low')
    )
)

# Exercise 10A
major_size_solution = students.groupby('major').size()

# Exercise 10B
major_summary_solution = students.groupby('major').agg(
    avg_score=('score', 'mean'),
    avg_attendance=('attendance', 'mean'),
    max_age=('age', 'max')
)

# Exercise 11A
merged_scores_solution = students[['student_id', 'name', 'major']].drop_duplicates().merge(
    scores_detail,
    how='inner',
    on='student_id'
)

# Exercise 11B
students_with_faculty_solution = students.merge(department_info, how='left', on='major')

# Exercise 12A
daily_sales_window_solution = daily_sales.assign(
    rolling_mean_3=daily_sales['sales'].rolling(3).mean(),
    cumulative_sales=daily_sales['sales'].cumsum()
)

# Exercise 14
practice_result_solution = (
    students
    .drop_duplicates()
    .loc[lambda df: df['score'] >= 80, ['name', 'major', 'score']]
    .sort_values('score', ascending=False)
)

# Exercise 15
challenge_result_solution = (
    students
    .drop_duplicates()
    .dropna(subset=['age'])
    .loc[lambda df: df['major'] == 'IT', ['name', 'age', 'score']]
    .sort_values('age')
)

print('products_solution')
display(products_solution)

print('scores_solution')
display(scores_solution)

print('high_score_students_solution')
display(high_score_students_solution)

print('target_students_solution')
display(target_students_solution)

print('selected_columns_solution')
display(selected_columns_solution)

print('regex_columns_solution')
display(regex_columns_solution)

print('sorted_students_solution')
display(sorted_students_solution)

print('renamed_students_solution')
display(renamed_students_solution)

print('sales_long_solution')
display(sales_long_solution)

print('sales_wide_again_solution')
display(sales_wide_again_solution)

print('missing_age_rows_solution')
display(missing_age_rows_solution)

print('students_dropna_age_solution')
display(students_dropna_age_solution)

print('students_fillna_age_solution')
display(students_fillna_age_solution)

print('scores_with_total_solution')
display(scores_with_total_solution)

print('students_band_solution')
display(students_band_solution)

print('major_size_solution')
display(major_size_solution)

print('major_summary_solution')
display(major_summary_solution)

print('merged_scores_solution')
display(merged_scores_solution)

print('students_with_faculty_solution')
display(students_with_faculty_solution)

print('daily_sales_window_solution')
display(daily_sales_window_solution)

print('practice_result_solution')
display(practice_result_solution)

print('challenge_result_solution')
display(challenge_result_solution)
```
